```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# blank white image
img = np.ones((500, 500, 3), dtype="uint8") * 255

cv2_imshow(img)
```

```python
line_img = img.copy()
cv2.line(line_img, (50, 50), (400, 50), (255, 0, 0), 3)

cv2_imshow(line_img)
```

```python
shape_img = img.copy()

cv2.rectangle(shape_img, (50, 100), (200, 250), (0, 255, 0), 3)
cv2.circle(shape_img, (300, 200), 50, (0, 0, 255), 3)

cv2_imshow(shape_img)
```

```python
multi_img = img.copy()

cv2.line(multi_img, (0, 0), (500, 500), (255, 0, 0), 2)
cv2.rectangle(multi_img, (50, 50), (150, 150), (0, 255, 0), 2)
cv2.circle(multi_img, (250, 250), 40, (0, 0, 255), -1)

cv2_imshow(multi_img)
```

```python
text_img = img.copy()

cv2.putText(text_img, "OpenCV", (150, 300),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (0, 0, 0), 2)

cv2_imshow(text_img)
```

```python
canvas = img.copy()

cv2.line(canvas, (50, 50), (450, 50), (255, 0, 0), 2)
cv2.rectangle(canvas, (50, 100), (200, 250), (0, 255, 0), 2)
cv2.circle(canvas, (350, 200), 50, (0, 0, 255), 2)
cv2.putText(canvas, "Canvas", (150, 400),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

cv2_imshow(canvas)
```
